```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\user\Documents\DEE\Analyst Portfolio\life expectancy.csv")
data
```

```python
#filtering using the region column to focus on sub-saharan africa

africa = ['Sub-Saharan Africa']
SSA = data[data['Region'] .isin(africa)]
SSA
```

```python
#Taking a look at the data

SSA.info()
```

```python
SSA.describe()
```

```python
SSA.isnull().sum()
```

```python
SSA.groupby('Year').agg({'Education Expenditure %' : ['mean']})
```

```python
#filling missing values in the data by columns
SSA
```

```python
SSA['Prevelance of Undernourishment'] = SSA.groupby('Year')['Prevelance of Undernourishment'].apply(lambda x: x.fillna(x.mean()))
SSA
```

```python
SSA['Health Expenditure %'] = SSA.groupby('Year')['Health Expenditure %'].apply(lambda x: x.fillna(x.mean()))
SSA['Education Expenditure %'] = SSA.groupby('Year')['Education Expenditure %'].apply(lambda x: x.fillna(x.mean()))
SSA['Unemployment'] = SSA.groupby('Year')['Unemployment'].apply(lambda x: x.fillna(x.mean()))
SSA['Sanitation'] = SSA.groupby('Year')['Sanitation'].apply(lambda x: x.fillna(x.mean()))
SSA['Corruption'] = SSA['Corruption'].fillna(0)
```
